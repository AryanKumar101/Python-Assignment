# Python Assignment - Working With Events

In this question, you are given 4 text files. 
These files contain timestamps, which correspond to the times of some 'events'. 

Your first task is to find out how many events are present in each file.

Consider only those events which have a positive value for timestamp


```python
import numpy as np
import matplotlib.pyplot as plt


eventQ1 = np.loadtxt("event_Q1.txt", delimiter=',')
eventQ2 = np.loadtxt("event_Q2.txt", delimiter=',')
eventQ3 = np.loadtxt("event_Q3.txt", delimiter=',')
eventQ4 = np.loadtxt("event_Q4.txt", delimiter=',')

events = [eventQ1, eventQ2, eventQ3, eventQ4]

print(f"No. of events in Q1: {len(eventQ1[np.where(eventQ1 > 0)])}")
print(f"No. of events in Q2: {len(eventQ2[np.where(eventQ2 > 0)])}")
print(f"No. of events in Q3: {len(eventQ3[np.where(eventQ3 > 0)])}")
print(f"No. of events in Q4: {len(eventQ4[np.where(eventQ4 > 0)])}")

```

After this, for each file, make a plot of events per second in that file.
You dont have to find the average rate, instead for each second 0-1, 1-2, 2-3, .. 54-55, find the number of events in each second and plot it with respect to time.

Plot these 4 separately, as well as in a single overlapping plot.
Make sure to give proper titles, labels and legends.

```python
labels = ["EventQ1", "EventQ2", "EventQ3", "EventQ4"]
UniversalCounter = []
for event in events:
    eventCount = []
    for i in range(int(max(event))):
        eventCount.append(sum(np.logical_and(event > i, event < i+1)))  #Here, sum returns the no. of "True" values
    UniversalCounter.append(eventCount)
    plt.figure()
    plt.plot(range(int(max(event))), eventCount)
    plt.xlim(0, 55)
    plt.xlabel("Time(s)")
    plt.ylabel("Frequency")
    plt.title("Event Frequency Plot")
    plt.grid()

for event in events:
    eventCount = []
    for i in range(int(max(event))):
        eventCount.append(sum(np.logical_and(event > i, event < i+1)))  #Here, sum returns the no. of "True" values
    
    plt.plot(range(int(max(event))), eventCount, label = "EventQ{}".format(events.index(event)+1))
    plt.xlim(0, 55)
    plt.xlabel("Time(s)")
    plt.ylabel("Frequency")
    plt.title("Event Frequency Plot")
    plt.legend()
    plt.grid()

```

Now for the final part of this question. We can see that the event frequency varies with time. Count how many bins are there, where atleast 2 of the files have more than 140 events in that bin, simultaneously.

```python
TotalBinCount = 0
for t in range(max(len(x) for x in UniversalCounter)):
        binCount = 0
        for i in range(4):
            try:
                if UniversalCounter[i][t] > 140:
                    binCount += 1
            except:
                 continue
        if binCount > 2:
            TotalBinCount += 1
print(TotalBinCount)
```
